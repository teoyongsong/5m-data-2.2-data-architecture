```python
from google.cloud import storage

def list_first_10_objects():
    # Initialize the client
    client = storage.Client.create_anonymous_client()

    # Get the bucket
    bucket = client.bucket("gcp-public-data-landsat")

    # List blobs (objects) in the bucket
    blobs = client.list_blobs(bucket)

    # Print the first 10 blobs with their sizes
    for i, blob in enumerate(blobs):
        if i >= 10:
            break
        print(f"Name: {blob.name}, Size: {blob.size} bytes")

if __name__ == "__main__":
    list_first_10_objects()
```
